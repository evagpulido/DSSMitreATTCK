```python
import numpy as np
import pandas as pd
```

LIMPIAR MONDAY

```python
df_monday = pd.read_csv('Monday-WorkingHours.pcap_ISCX.csv')
```

```python
df_monday.head()
```

```python
print(df_monday.columns)
```

Nos quedamos solo con las columnas que nos interesan, en especial nos interesa la columna Label

```python
df_monday = df_monday[[ 'Flow ID', ' Timestamp', ' Label']]
```

```python
df_monday.head()
```

Elimino los benignos

```python
df_monday = df_monday[~df_monday[' Label'].str.contains('BENIGN')]
```

```python
df_monday.head()
```

Ya tenemos un dataframe solo con ataques

Ahora quiero guardar el DataFrame en un nuevo archivo CSV.


```python
df_monday.to_csv('LIMPIOS/Monday-WorkingHours.pcap_ISCX.csv', index=False)  # el parámetro 'index=False' evita que se incluya el índice en el archivo
```
